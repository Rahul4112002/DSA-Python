# Print N natural Numbers

```python
def printN(n):
    if n > 0:
        printN(n-1)
        print(n,end=' ')
printN(10)
```

# Print N natural Numbers in Reverse Order

```python
def printN(n):
    if n > 0:
        print(n , end = ' ')
        printN(n-1)
printN(10)
```

# Print First 10 Odd numbers

```python
def odd(n):
    if n > 0:
        odd(n-1)
        print(2*n-1,end= ' ')
odd(10)
        
```

# Print First 10 even numbers

```python
def even(n):
    if n > 0:
        even(n-1)
        print(2*n,end= ' ')
even(10)
```

# Print First 10 Odd numbers reverse order

```python
def odd(n):
    if n > 0:
        print(2*n-1,end= ' ')
        odd(n-1)   
odd(10)
```

# Print Sum of N natural Numbers

```python
def sumN(n):
    if n == 0:
        return 0
    else:
        return n + sumN(n-1)
sumN(10)
```

# Print Sum of n ODD natural Numbers

```python
def sumOdd(n):
    if n == 0:
        return 0
    return 2*n-1 + sumOdd(n-1)
sumOdd(3)
```

# Print Sum of n Even natural Numbers

```python
def sumEven(n):
    if n == 0:
        return 0
    return 2*n + sumEven(n-1)
sumEven(3)
```

# Print Factorial of Number

```python
def factorial(n):
    if n == 0:
        return 1
    return n*factorial(n-1)
factorial(5)
```

# Print Sum of Square of n Natural Number

```python
def sumSqr(n):
    if n == 0:
        return 0
    return n**2 + sumSqr(n-1)
sumSqr(3)
```
