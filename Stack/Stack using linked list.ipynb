# Stack using linked list

```python
class Node:
    def __init__(self,item=None,next = None):
        self.item = item
        self.next = next
class Stack:
    def __init__(self,start = None):
        self.start = start
        self.item_count = 0
    def is_empty(self):
        return self.start == None
    def push(self,data):
        n = Node(data,self.start)
        self.start = n
        self.item_count += 1
    def pop(self):
        if not self.is_empty():
            data = self.start.item
            self.start  = self.start.next
            self.item_count -= 1
            return data
        else:
            raise IndexError('Stack Underflow')
    def peek(self):
        if not self.is_empty():
            return self.start.item
        else:
            raise IndexError('Stack Underflow')
    def size(self):
        return self.item_count
    def print_stack(self):
        current = self.start
        while current:
            print(current.item, end=" ")
            current = current.next
```

```python
s = Stack()
```

```python
s.push(10)
s.push(20)
s.push(30)
```

```python
print(s.peek())
```

```python
print(s.pop())
```

```python
print(s.peek())
```

```python
print(s.size())
```

```python
print(s.print_stack())
```
